```python
soup = BeautifulSoup(response.text, 'html')
```

<div class="alert alert-block alert-warning">
<b>Cvičení:</b> <br>Vyextrahuj všechny odkazy z části, kterou jsi získal(a) v předchozí fázi.

```python
links = []
for element in elements:
    link_element = element.find("a")
    link = link_element.get("href")
    links.append(link)
```

```python
links
```

```python
urls = []
for link in links:
    url = domain + link
    urls.append(url)
    
urls
```

```python
kocky = []

for url in urls:
    response = requests.get(url)

    if response.status_code == 200:
        html_content = response.text

    else:
        print(f"Chyba při načítání stránky. HTTP kód: {response.status_code}")
        
    soup = BeautifulSoup(response.text, 'html')
    
    charakteristika = {}
        
    jmeno = soup.find("title").text
    charakteristika["title"] = jmeno
    
    elements = soup.find_all('div', class_='tags')
    for element in elements:
        if len(element.text.strip().split("\t")) == 1:
            pass
        else:
            key, value = element.text.replace("\n", " ").strip().split("\t")
            key = " ".join(key.split())
            value = " ".join(value.split())
            charakteristika[key] = value
        #print(element.find('h4').find('a').text.strip())

    print(charakteristika)
    kocky.append(charakteristika)
    sleep(5)
```

```python
url = "https://www.kravihora-brno.cz/venkovni-bazeny"
response = requests.get(url)

if response.status_code == 200:
    html_content = response.text
    print(html_content)
else:
    print(f"Chyba při načítání stránky. HTTP kód: {response.status_code}")
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'lxml')
    
    # Představme si, že chcete najít první <h1> element na stránce:
    h1_tag = soup.find('h1')
    
    if h1_tag:
        print(h1_tag.text)  # Vytiskne text uvnitř <h1> elementu
        selector = "#block-kravihora-content > article > section.border-left.border-right.border-top.info.margin-bottom-0.layout.section_layout.layout--one-column-extra > div > div > div > div.block.block-layout-builder.block-field-blocknodefacilitiesfield-address > div > div.field__item > p:nth-child(3) > strong"
    
        element = soup.select_one(selector)
        print(element.text)
    else:
        print("Element <h1> nebyl na stránce nalezen.")
else:
    print(f"Chyba při načítání stránky. HTTP kód: {response.status_code}")
```
